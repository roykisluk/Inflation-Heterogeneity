# Setup and Data

```python
from inflation_analysis import tri_grouping
import matplotlib.pyplot as plt

# Parameters
start_year = 2021
end_year = 2022
data_folder="/Users/roykisluk/Downloads/Consumer_Expenditure_Survey/"
years=range(start_year, end_year+1)

demo, income, ses, total_mmb = tri_grouping(start_year, end_year, cex_data_folder = data_folder)
```

# Plotting

## Demographic groups

```python
# Count rows for each group
group_counts = {group: sum(len(data[year]) for year in data) for group, data in demo.items()}

# Plot
plt.figure(figsize=(10, 6))
plt.bar(group_counts.keys(), group_counts.values(), color='skyblue')
plt.xlabel('Groups')
plt.ylabel('Number of Observations')
plt.title('Number of Observations for Each Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Income groups

```python
# Count rows for each group
group_counts = {group: sum(len(data[year]) for year in data) for group, data in income.items()}

# Plot
plt.figure(figsize=(10, 6))
plt.bar(group_counts.keys(), group_counts.values(), color='skyblue')
plt.xlabel('Groups')
plt.ylabel('Number of Observations')
plt.title('Number of Observations for Each Group')
plt.xticks(ticks=range(1,len(group_counts)+1), labels=group_counts.keys(), ha='right')
plt.tight_layout()
plt.show()
```
